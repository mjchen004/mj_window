```python
import requests
from requests import Response
from pydantic import ValidationError
from requests import JSONDecodeError

def download_json()->dict[any]:

    url = 'https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json'

    # get json data from the url
    try:
        res:Response = requests.get(url)
    except Exception: 
        raise Exception("connection failed!")
    else: 
        if res.status_code == 200: 
            try: 
                all_data:dict[any] = res.json()
                return all_data

            except JSONDecodeError:
                raise Exception("cannot decode text into json!")
        else:
            raise Exception("Download failed!")
```

```python
try:
    all_data:dict[any] = download_json()
except Exception as error:
    print(error)
```

```python
from pydantic import BaseModel, Field, RootModel #basemodel -> dictionary , rootmodel -> list , field -> string

class Information(BaseModel): 
    sna: str
    sarea: str
    mday: str
    ar: str
    act: int
    updateTime: str
    total: int
    bikesavailable: int = Field(alias='available_rent_bikes')
    bikesreturn: int = Field (alias='available_return_bikes')
    latitude: float
    longitude: float

class Youbike(RootModel): 
    root: list[Information]

# valiade + dump to dict
youbike:Youbike = Youbike.model_validate(all_data)
data = youbike.model_dump()
data
```
